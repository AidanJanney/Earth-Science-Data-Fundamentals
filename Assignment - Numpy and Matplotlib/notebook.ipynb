```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1 Creating and Manipulating Arrays

### 1.1 Create two 2D arrays representing coordinates x,y on the cartesian plane

```python
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
xx, yy = np.meshgrid(x, y)

```

### 1.2. Visualize each 2D array using pcolormesh

```python
mesh_xx = plt.pcolormesh(xx)
plt.colorbar(mesh_xx)
```

```python
mesh_yy = plt.pcolormesh(yy)
plt.colorbar(mesh_yy)
```

### 1.3 Convert to Polar  Coordinates

```python
rr = np.sqrt(xx**2 + yy**2)
phiphi = np.arctan2(yy,xx)
```

```python
mesh_rr = plt.pcolormesh(rr)
plt.colorbar(mesh_rr)
```

```python
mesh_phiphi = plt.pcolormesh(phiphi)
plt.colorbar(mesh_phiphi)
```
